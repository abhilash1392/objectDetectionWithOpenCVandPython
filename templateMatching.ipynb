```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
full=cv2.imread('D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\sammy.jpg')
full=cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face=cv2.imread('D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\sammy_face.jpg')
face=cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
face.shape
```

```python
height,width,channels=face.shape
```

```python
print(height,width,channels)
```

```python
# All the 6 methods for comparison in a list
# Note how we are using strings, later on we'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    fullCopy=full
    method=eval(m)
    res=cv2.matchTemplate(fullCopy,face,method)
    minVal,maxVal,minLoc,maxLoc=cv2.minMaxLoc(res)
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        topLeft=minLoc
    else:
        topLeft=maxLoc

    bottomRight=(topLeft[0]+width,topLeft[1]+height)
    cv2.rectangle(fullCopy,topLeft,bottomRight,(255,0,0),10)
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEATMAP')
    plt.subplot(122)
    plt.imshow(fullCopy)
    plt.title('Detected Face')
    plt.suptitle(m)

    plt.show()
    print('\n')
    print('\n')
```
