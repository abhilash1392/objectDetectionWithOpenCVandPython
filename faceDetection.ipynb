```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia=cv2.imread(r'/home/abhilash/Coding/computervision/objectDetectionWithOpenCVandPython/DATA/Nadia_Murad.jpg',0)
solvay=cv2.imread(r'/home/abhilash/Coding/computervision/objectDetectionWithOpenCVandPython/DATA/solvay_conference.jpg',0)
dennis=cv2.imread(r'/home/abhilash/Coding/computervision/objectDetectionWithOpenCVandPython/DATA/Denis_Mukwege.jpg',0)
```

```python
faceCascade=cv2.CascadeClassifier('/home/abhilash/Coding/computervision/objectDetectionWithOpenCVandPython/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detectFace(img):
    faceImage=img.copy()
    faceRects=faceCascade.detectMultiScale(faceImage)
    for (x,y,w,h) in faceRects:
        cv2.rectangle(faceImage,(x,y),(x+w,y+h),(255,255,255),10)

    return faceImage
```

```python
resultDennis=detectFace(dennis)
plt.imshow(resultDennis,cmap='gray');
```

```python
resultNadia=detectFace(nadia)
plt.imshow(resultNadia,cmap='gray');
```

```python
resultSolvay=detectFace(solvay)
plt.imshow(resultSolvay,cmap='gray');
```

```python
def adjDetectFace(img):
    faceImage=img.copy()
    faceRects=faceCascade.detectMultiScale(faceImage,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in faceRects:
        cv2.rectangle(faceImage,(x,y),(x+w,y+h),(255,255,255),10)

    return faceImage
```

```python
resultSolvayNew=adjDetectFace(solvay)
plt.imshow(resultSolvayNew,cmap='gray');
```

```python
cap=cv2.VideoCapture(0)

while True:

    ret,frame=cap.read(0)

    frame=detectFace(frame)

    cv2.imshow('Video Face Detect',frame)

    k=cv2.waitKey(1)

    if k ==27:
        break

cap.release()
cv2.destroyAllWindows()
```
