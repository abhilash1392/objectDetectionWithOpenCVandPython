```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
img=cv2.imread(r'D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\internal_external.png',0)
plt.imshow(img,cmap='gray')
```

```python
contours,hierarchy=cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
print(len(contours))
print((hierarchy))
```

```python
externalContours=np.zeros(shape=img.shape)
```

```python
externalContours.shape
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3]==-1:
        cv2.drawContours(externalContours,contours,i,255,-1)
```

```python
plt.imshow(externalContours,cmap='gray')
```

```python
internalContours=np.zeros(shape=img.shape)
for i in range(len(contours)):
    if hierarchy[0][i][3]!=-1:
        cv2.drawContours(internalContours,contours,i,255,-1)
```

```python
plt.imshow(internalContours,cmap='gray')
```
