```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
# Opening and reading file Flat Chess board
path=r'D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\flat_chessboard.png'
flatChess=cv2.imread(path)
flatChess=cv2.cvtColor(flatChess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flatChess)
```

```python
grayChess=cv2.cvtColor(flatChess,cv2.COLOR_RGB2GRAY)
plt.imshow(grayChess,cmap='gray')
```

```python
realPath=r'D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\real_chessboard.jpg'
realChess=cv2.imread(realPath)
realChess=cv2.cvtColor(realChess,cv2.COLOR_BGR2RGB)
plt.imshow(realChess)
```

```python
gray=np.float32(grayChess)
```

```python
dst=cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)
dst=cv2.dilate(dst,None)
```

```python
flatChess[dst>0.1*dst.max()]=[255,0,0]
```

```python
plt.imshow(flatChess)
```

```python
grayRealChess=cv2.cvtColor(realChess,cv2.COLOR_RGB2GRAY)
plt.imshow(grayRealChess,cmap='gray')
```

```python
gray=np.float32(grayRealChess)
dst=cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)
dst=cv2.dilate(dst,None)
realChess[dst>0.1*dst.max()]=[255,0,0]
plt.imshow(realChess);
```

```python
#Shi-Tomasi Algorithm
path=r'D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\flat_chessboard.png'
flatChess=cv2.imread(path)
flatChess=cv2.cvtColor(flatChess,cv2.COLOR_BGR2RGB)

realPath=r'D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\real_chessboard.jpg'
realChess=cv2.imread(realPath)
realChess=cv2.cvtColor(realChess,cv2.COLOR_BGR2RGB)

grayChess=cv2.cvtColor(flatChess,cv2.COLOR_RGB2GRAY)

grayRealChess=cv2.cvtColor(realChess,cv2.COLOR_RGB2GRAY)
```

```python
corners=cv2.goodFeaturesToTrack(grayChess,80,0.01,10)
corners=np.int0(corners)
```

```python
for i in corners:
    x,y=i.ravel()
    cv2.circle(flatChess,(x,y),3,255,-1)

plt.imshow(flatChess)
```

```python
corners=cv2.goodFeaturesToTrack(grayRealChess,120,0.01,10)
corners=np.int0(corners)

for i in corners:
    x,y=i.ravel()
    cv2.circle(realChess,(x,y),3,255,-1)

plt.imshow(realChess)
```
