```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
img=cv2.imread('D:\Coding\OpenCVTutorial\code\objectDetectionWithOpenCVandPython\DATA\sammy_face.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

```python
# Using Canny Edge detection algorithm
medianVal=np.median(img)
print(medianVal)
```

```python
# Setting Limit for Lower and Upper Thresholds
lower=int(max(0,0.7*medianVal))
print(lower)
upper=int(min(255,1.3*medianVal))
print(upper)
```

```python
edges=cv2.Canny(img,lower,upper)
plt.imshow(edges)
```

```python
# As it is obeserved that it is not performing very well we will blur the image
blurImg=cv2.blur(img,ksize=(7,7))
plt.imshow(blurImg)
```

```python
edges=cv2.Canny(blurImg,lower,upper+25)
plt.imshow(edges)
```
